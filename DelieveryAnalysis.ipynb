## Batsman Analysis

```python
#all batsman name
man = balls['batsman'].unique()
```

```python
man
```

```python
# Total run by batsman
runs = balls.groupby('batsman').sum()
runs = runs.sort_values(by=['batsman_runs', 'batsman'], ascending=False)
runs = runs.drop(['match_id', 'inning', 'over', 'extra_runs','total_runs'], axis=1)
runs
```

```python
# total runs by batsman in every match and centuries
run=pd.DataFrame([balls.match_id, balls.inning, balls.batsman, balls.batsman_runs]).transpose()
```

```python
run
```

```python
#run=pd.DataFrame([run.match_id, run.inning, run.batsman, run.batsman_run])
run = run.groupby(['match_id','inning', 'batsman']).sum()
run
```

```python
centuries = run[run['batsman_runs']>=100].sort_values(by=['batsman'])
centuries.head(50)
```
